```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
```

```python
'''since we have multiple files we can write a function to import all the files and since the hae the same columns 
concat them at the same time'''
#importin the modules
import os
import glob

#path where all the csv files are located
path =r'C:\Users\trevour\Desktop\trashout omdena\TrashOut_data_export'

A_files = glob.glob(os.path.join(path, "*.csv"))     
DataFrames = (pd.read_csv(f,sep=';') for f in A_files)
A_data   = pd.concat(DataFrames, ignore_index=True)
```

```python
A_data.shape
#we have a total record of 56,109 from allover the world
```

```python
#dropping the columns that are deemed irrelevant or are of least importance as discussed
A_data = A_data.drop(['continent','aa2','aa3','id','url','anonymous','note','image'],axis=1)
#view our updated dataframe
A_data.head()
```

```python
# for easier viewing and understanding of the data w hae we could set the index as the country
A_data = A_data.set_index(['country','locality'])
A_data.head()
```

```python
#overview of the data so far
A_data.info()
```

```python
#some of the data contains a wholesum of NA values
A_data.isnull().sum()
```

```python
#we definitely need to handle the missing data especially the sublocality and accessibulity
A_data = A_data.drop(['accessibility','subLocality'],axis=1)
```

```python
A_data.head()
```

```python
#we can now drop the remaining na 
A_data = A_data.dropna()
```

```python
A_data.info()
```

```python
#lets convert our date from an object type datetime
from datetime import datetime
A_data['updateTime']=A_data['updateTime'].map(lambda x :datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%fZ") )
A_data['updateTime']
```
